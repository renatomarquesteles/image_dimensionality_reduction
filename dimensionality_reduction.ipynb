```python
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Upload image
uploaded = files.upload()
color_image_path = list(uploaded.keys())[0]

# Load image
image = Image.open(color_image_path)

# Resize image to 640x480
image_resized = image.resize((640, 480))

# Show image
plt.imshow(image_resized, cmap=None)
plt.axis('off')
plt.show()
```

```python
# Convert to Gray
image_gray = image_resized.convert('L')
# Show image
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Convert to binary (black and white)
threshold = 127
image_binary = image_gray.point(lambda p: 255 if p > threshold else 0)
# Show image
plt.imshow(image_binary, cmap='gray')
plt.axis('off')
plt.show()
```
